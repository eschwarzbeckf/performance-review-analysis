```python
# Install necessary library
!pip install transformers

# Import zero-shot classification pipeline
from transformers import pipeline

# Load the zero-shot classifier
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Your example texts
texts = [
    "Women are not good at math.",
    "Men are always better leaders.",
    "People of all races deserve equal rights.",
    "She is an excellent scientist.",
    "Black people are more likely to commit crimes.",
    "Maria is very pleasant and gets along well with her colleagues. She always tries her best and is very polite. While she may not be the most technical person on the team, she brings a nurturing presence to the group. She sometimes hesitates to take the lead on complex projects, but that’s understandable. Compared to others, she tends to need more direction and reassurance. Her attention to detail is good, but she can become emotional under pressure, which can affect team morale. She should continue to work on being more assertive and independent. Also, while her communication is soft-spoken, it would help her to be more vocal like some of the male team members."
]

# Hypothesis (bias detection)
hypothesis = "This text is biased."

# Run the classifier
for i, text in enumerate(texts):
    result = classifier(text, candidate_labels=["biased", "unbiased"])
    label = result['labels'][0]
    score = result['scores'][0]
    print(f"--- Example {i+1} ---")
    print(f"Text: {text}")
    print(f"Prediction: {label} | Confidence: {score:.4f}\n")
```
